```python
from rdkit import Chem

canonical_smi = "CN1C=NC2=C1C(=O)N(C(=O)N2C)C"
mol = Chem.MolFromSmiles(canonical_smi)

# 2. Generate 5 different non-canonical representations
print("--- Non-Canonical SMILES ---")
for i in range(1000):
    # Setting doRandom=True skips the canonical ordering algorithm
    non_canonical_smi = Chem.MolToSmiles(mol, doRandom=True)
    print(f"{non_canonical_smi}")

# 3. Canonical SMILES (always the same)
print("\n--- Canonical SMILES ---")
print(Chem.MolToSmiles(mol))
```
